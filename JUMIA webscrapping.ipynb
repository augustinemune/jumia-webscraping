```python
url = "https://www.jumia.co.ke/home-audio-speakers/"
```

```python
r = requests.get(url)
r.status_code
```

```python
html_strg = r.text
html = HTML(html=html_strg)
```

```python
from bs4 import BeautifulSoup
from csv import writer
```

```python
soup = BeautifulSoup(html_strg, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all = soup.find_all(class_='_fb')
print(len(all))
```

```python
for i in all:
  title = soup.select('.name')
  price = soup.select('.prc')
  print(title)
```

```python
with open('home-audio-speakers.csv', 'w') as filename_csv:
    csv_writer = writer(filename_csv)
    header = ['Title', 'Price', 'TV Brad']
    csv_writer.writerow(header)
    
    ff = soup.find_all(class_='_fb')
    for article in ff:
        title = article.find(class_='name').get_text()
        price = article.find(class_='prc').get_text()

        def Convert(string): 
            li = list(string.split(" ")) 
            return li 

        # Driver code     
        str1 = title
        Brad = Convert(str1)[0]

        csv_writer.writerow([title, price, Brad])
```
